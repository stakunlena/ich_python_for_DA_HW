1. Основываясь на датасете iris из seaborn, постройте следующий график.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Загружаем датасет iris
iris = sns.load_dataset("iris")

# Фильтруем по видам
setosa = iris[iris["species"] == "setosa"]
versicolor = iris[iris["species"] == "versicolor"]
virginica = iris[iris["species"] == "virginica"]

# Строим вручную три гистограммы с разными bins
plt.figure(figsize=(8, 5))

sns.histplot(setosa["petal_length"], bins=20, color="#8ebad9ff", edgecolor=None, label="setosa", alpha=0.75)
sns.histplot(versicolor["petal_length"], bins=20, color="#ffbe86ff", edgecolor=None, label="versicolor", alpha=0.75)
sns.histplot(virginica["petal_length"], bins=20, color="#95cf95ff", edgecolor=None, label="virginica", alpha=0.75)

plt.title("Iris Species Distribution by Petal Length\n(разное количество bins для разных видов)")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Count")
plt.legend()
plt.show()
```
