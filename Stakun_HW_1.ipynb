**1.1  Сложение двух массивов:**

 Напишите программу, которая:
1. Запрашивает у пользователя ввод двух одномерных массивов одинаковой длины.
2. Массивы вводятся в виде чисел, разделенных пробелами.
3. Преобразует введенные строки в массивы NumPy.
4. Складывает два массива поэлементно.
5. Выводит результат сложения на экран.

```python
import numpy as np

list1 = list(map(int, input("Введите первый список из 3 элементов (через пробел): ").split()))
list2 = list(map(int, input("Введите второй список из 3 элементов (через пробел): ").split()))

array1 = np.array(list1)
array2 = np.array(list2)

print("Результат сложения:", array1 + array2)
```

```python
#2 вариант
array1 = np.array(list(map(int, input("Введите первый список из 3 элементов (через пробел): ").split())))
array2 = np.array(list(map(int, input("Введите первый список из 3 элементов (через пробел): ").split())))

print("Результат сложения:", array1+array2)
```

```python
#3 вариант
array1, array2 = (np.array(list(map(int, input(f"Введите список {i+1} чисел через пробел: ").split()))) for i in range(2))
if array1.shape != array2.shape:
    print("Ошибка: массивы должны быть одинаковой длины.")
else:
    print("Результат сложения:", array1 + array2)
```

**1.2 Поэлементное умножение массивов**

   Напишите программу, которая:
1. Запрашивает у пользователя размер массивов N.
2. Генерирует два одномерных массива длины N с случайными целыми числами от 1 до 10.
3. Выводит оба сгенерированных массива на экран.
4. Выполняет поэлементное умножение массивов.
5.Выводит результат умножения на экран.

```python
N = int(input("Введите размер массивов (целое число): "))

array1, array2 = (np.random.randint(1, 11, size=N) for _ in range(2))

print("Массив 1:", array1)
print("Массив 2:", array2)

result = array1 * array2

print("Результат поэлементного умножения:", result)
```

**1.3 Создание и модификация массива**
   
   Напишите программу, которая:

1. Запрашивает у пользователя количество элементов массива N.
2. Генерирует одномерный массив длины N, заполненный случайными числами от 0 до 100.
3. Выводит сгенерированный массив на экран.
4. Запрашивает у пользователя число X.
5. Умножает каждый элемент массива на число X.
6. Выводит измененный массив на экран.

```python
n = int(input("Введите количество элементов массива: "))

array1 = np.random.randint(0, 101, size=n)

print("Массив 1:", array1)

x = int(input("Введите число для умножения: "))

result = array1 * x

print("Результат:", result)
```
