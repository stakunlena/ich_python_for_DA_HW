**1. Использование пользовательских индексов в Series:**

Напишите программу, которая:
1. Запрашивает у пользователя ввод чисел, разделенных пробелами.
1. Преобразует введенные данные в объект Series, используя в качестве индексов буквы алфавита начиная с 'a'.
1. Выводит созданный объект Series на экран.


```python
import pandas as pd

i = list('abcd') # задаем в качестве индекса буквы латинского алфавита i == ['a', 'b', 'c', 'd']

# просим ввести 4 числа и создаем из них объект Series с идексом i (буквы алфавита)
s = pd.Series(map(int, input("Введите 4 числа (через пробел) и нажмите Enter: ").split()), i)

s # выводим созданный объект Series на экран.
```

**2. Создание и индексация DataFrame:**

Напишите программу, которая:
1. Запрашивает у пользователя данные для создания DataFrame. Данные включают в себя имена столбцов и строки значений, вводимые через пробел.
1. Создает DataFrame с использованием введенных данных.
1. Выводит созданный DataFrame на экран.
1. Запрашивает у пользователя имя столбца, по которому необходимо отфильтровать данные.
1. Выводит значения указанного столбца.

```python
import pandas as pd

# 1. Вводим имена столбцов и данные построчно, пока не введен end
columns = input("Введите имена столбцов: ").split()

data = []
while True:
    row = input("Введите данные (для окончания введите 'end'): ")
    if row.lower() == 'end':
        break
    values = row.split()
    if len(values) != len(columns):
        print("Количество значений не совпадает с количеством столбцов. Попробуйте снова.")
        continue
    data.append(values)

# 2. Создаем DataFrame
df = pd.DataFrame(data, columns=columns)

# 3. Выводим созданный DataFrame
print("\nСозданный DataFrame:")
print(df)

# 4. Запрашиваем имя столбца для фильтрации
column_name = input("\nВведите имя столбца для отображения: ")

# 5. Проверяем наличие столбца и выводим значения
if column_name in df.columns:
    print(f"\n{column_name}")
    print(df[column_name])
else:
    print("Такого столбца нет в DataFrame.")
```
