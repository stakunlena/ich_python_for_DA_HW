**1. Заполнение пустых значений**

Загрузите датасет titanic из библиотеки seaborn, определите столбцы с пустыми значениями и заполните их любым из возможных методов.

```python
import seaborn as sns
import pandas as pd

# Загружаем датасет
titanic = sns.load_dataset("titanic")

# Проверяем пропуски
print("Количество пропусков до заполнения:")
print(titanic.isnull().sum())

# Заполняем пропуски корректно
for column in titanic.columns:
    if titanic[column].dtype in ["float64", "int64"]:
        titanic[column] = titanic[column].fillna(titanic[column].mean())
    else:
        titanic[column] = titanic[column].fillna(titanic[column].mode()[0])

# Проверяем результат
print("\nКоличество пропусков после заполнения:")
print(titanic.isnull().sum())
```

**2. Кодирование категориальных переменных**

Загрузите датасет titanic из библиотеки seaborn, определите категориальную переменную и закодируйте её.

```python
import seaborn as sns
import pandas as pd

# 1. Загружаем датасет Titanic
titanic = sns.load_dataset("titanic")

# 2. Смотрим первые строки, чтобы понять структуру
print("Первые строки датафрейма:")
print(titanic.head())

# 3. Определяем категориальные переменные
print("\nТипы данных в датафрейме:")
print(titanic.dtypes)

# Например, возьмём категориальный столбец 'sex'
print("\nУникальные значения в столбце 'sex':")
print(titanic["sex"].unique())

# 4. Кодируем категориальную переменную 'sex' в числовую форму
titanic["sex_code"] = titanic["sex"].map({"male": 0, "female": 1})

# 5. Проверяем результат
print("\nПосле кодирования:")
print(titanic[["sex", "sex_code"]].head())
```
