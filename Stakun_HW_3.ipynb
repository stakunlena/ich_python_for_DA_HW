**1. Обратная матрица и определитель:**

Напишите программу, которая:
1. Запрашивает у пользователя размер квадратной матрицы N.
1. Генерирует квадратную матрицу размером N x N, заполненную случайными числами от -10 до 10.
1. Выводит сгенерированную матрицу на экран.
1. Вычисляет и выводит определитель матрицы.
1. Если определитель не равен нулю, программа должна вычислить и вывести обратную матрицу. В противном случае, программа должна сообщить, что обратной матрицы не существует.

```python
import numpy as np
try:
    N = int(input("Введите размер матрицы N: "))

    if N <= 0:
        print("Размер матрицы должен быть положительным числом.")
    else:
        # Генерируем матрицу NxN со случайными целыми числами от -10 до 10
        matrix = np.random.randint(-10, 11, size=(N, N))

        print("\nСгенерированная матрица:")
        print(matrix)

        try:
            # Вычисляем определитель
            determinant = np.linalg.det(matrix)
            print("\nОпределитель матрицы:", round(determinant, 2))

            # Проверяем, существует ли обратная матрица
            if determinant != 0:
                inverse_matrix = np.linalg.inv(matrix)

                print("\nОбратная матрица:")
                print(np.round(inverse_matrix, 2))  # округляем до 2 знаков
            else:
                print("\nОбратной матрицы не существует (определитель равен 0).")

        except np.linalg.LinAlgError:
           print("\nОшибка при вычислении обратной матрицы.")

except ValueError:
    # Обработка ошибки, если пользователь ввёл не число
    print("Ошибка: введите целое число.")
```
