Задание 1.

С помощью библиотеки scipy.stats выполните независимый t-тест на наборе данных tips из seaborn. Сравните средний общий счет между курящими и некурящими.

```python
import seaborn as sns
from scipy import stats

# 1. Загружаем набор данных
tips = sns.load_dataset("tips")

# 2. Разделяем данные на две группы: курящие и некурящие
smokers = tips[tips["smoker"] == "Yes"]["total_bill"]
nonsmokers = tips[tips["smoker"] == "No"]["total_bill"]

# Постановка гипотез:
# H0 - средние счета одинаковы
# H1 - средние счета различаются

# 3. Выполняем независимый t-тест
t_statistic, p_value = stats.ttest_ind(smokers, nonsmokers, equal_var=False)

print(f"T-статистика: {t_statistic}")
print(f"P-value: {p_value}")

# 4. Интерпретация результата
alpha = 0.05 # уровень значимости

if p_value < alpha:
    print("Отклоняем H0. Средние счета между курящими и некурящими значимо различаются.")
else:
    print("Не отклоняем H0. Нет статистически значимого различия между группами.")
```

Задание 2.

С помощью библиотеки statsmodels выполните одновыборочный z-тест для набора данных по бриллиантам из seaborn. Проверьте, значительно ли отличается средняя цена бриллиантов от $4000.

```python
import seaborn as sns
import numpy as np
from statsmodels.stats.weightstats import ztest

# 1. Загружаем набор данных
diamonds = sns.load_dataset("diamonds")

# 2. Извлекаем столбец "price"
price = diamonds["price"]

# Постановка гипотез:
# H0 - средняя цена бриллиантов = 4000
# H1 - средняя цена бриллиантов ≠ 4000

# 3. Выполняем одновыборочный Z-тест
# Проверяем H0: mean(price) = 4000
test_stat, p_value = ztest(price, value=4000)

# 4. Выводим результаты
print(f"Z-score: {test_stat:.4f}")
print(f"P-value: {p_value:.4e}")

# 5. Интерпретация
alpha = 0.05 # уровень значимости

if p_value < alpha:
    print("Отклоняем H0. Средняя цена бриллиантов статистически отличается от $4000.")
else:
    print("Не отклоняем H0. Нет статистически значимого отличия от $4000.")
```

Задание 3.

Используйте библиотеку scipy.stats для расчета коэффициента корреляции Пирсона между total_bill и tip в наборе данных tips.

```python
import seaborn as sns
from scipy.stats import pearsonr

# 1. Загружаем данные
tips = sns.load_dataset("tips")

# 2. Извлекаем нужные столбцы
x = tips["total_bill"]
y = tips["tip"]

# 3. Вычисляем коэффициент корреляции Пирсона
r_value, p_value = pearsonr(x, y)

# 4. Выводим результаты
print(f"Коэффициент корреляции Пирсона r = {r_value:.4f}")
print(f"P-value = {p_value:.4e}")

# Интерпретация результата
alpha = 0.05 # уровень значимости

if p_value < alpha:
    print("Связь статистически значима.")
else:
    print("Нет статистически значимой линейной связи.")
```
