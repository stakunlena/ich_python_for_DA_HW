### Импорт необходимых библиотек

```python
import plotly.express as px
import numpy as np
import pandas as pd
import seaborn as sns
```

### Импорт данных

```python
df =  px.data.gapminder()

df.head()
```

### Приложение Dash

```python
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import plotly.express as px

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Life Expectancy Distribution by Continent"),
    dcc.Dropdown(
        id='continent-dropdown',
        options=[{'label': i, 'value': i} for i in df['continent'].unique()],
        value='Africa' # Default value
    ),
    dcc.Graph(id='lifeExp-histogram')
])

@app.callback(
    Output('lifeExp-histogram', 'figure'),
    Input('continent-dropdown', 'value')
)
def update_histogram(selected_continent):
    filtered_df = df[df['continent'] == selected_continent]
    fig = px.histogram(filtered_df, x="lifeExp", title=f"Distribution of Life Expectancy in {selected_continent}")
    return fig

if __name__ == '__main__':
    app.run(debug=True, mode='inline')
```

### plotly.figure_factory

```python
import plotly.figure_factory as ff

fig = ff.create_distplot([df[df['continent'] == c]['lifeExp'] for c in df['continent'].unique()], df['continent'].unique(), bin_size=.25)
fig.update_layout(hovermode='closest', clickmode='event+select') # Add interactivity
fig.show()
```
