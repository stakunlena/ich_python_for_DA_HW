**1. Сводные таблицы и таблицы сопряженности**

В датафрейме ниже представлены результаты соревнований по электронным играм: для каждого игрока указана игра и набранное количество очков в нескольких попытках.
```
data = {

    'Name': ['Анна', 'Антон', 'Мария', 'Макс', 'Макс', 'Алекс', 'Антон', 'Мария', 'Макс', 'Алекс', 'Антон', 'Анна'],
    'Game': ['тетрис', 'змейка', 'змейка', 'тетрис', 'тетрис', 'змейка', 'тетрис', 'тетрис', 'тетрис', 'змейка', 'змейка', 'тетрис'],
    'Points': [145, 2524, 343, 4212, 5212, 6421, 745, 842, 524, 3224, 212, 1214]      

}
```
Вам нужно узнать среднее количество очков каждого игрока по каждой игре. Используйте сводные таблицы или таблицы сопряженности.

````python
# Вариант решения с использованием сводной таблицы

import pandas as pd

# 1. Исходные данные
data = {
    'Name': ['Анна', 'Антон', 'Мария', 'Макс', 'Макс', 'Алекс', 'Антон', 'Мария', 'Макс', 'Алекс', 'Антон', 'Анна'],
    'Game': ['тетрис', 'змейка', 'змейка', 'тетрис', 'тетрис', 'змейка', 'тетрис', 'тетрис', 'тетрис', 'змейка', 'змейка', 'тетрис'],
    'Points': [145, 2524, 343, 4212, 5212, 6421, 745, 842, 524, 3224, 212, 1214]
}

# 2. Создаем DataFrame
df = pd.DataFrame(data)

# 3. Строим сводную таблицу: среднее количество очков по каждому игроку и игре
pivot_table = pd.pivot_table(df, values='Points', index='Name', columns='Game', aggfunc='mean')

# 4. Выводим исходные данные и сводную таблицу
print("Исходные данные:")
print(df)

print("\nСреднее количество очков каждого игрока по каждой игре:")
print(pivot_table)
````

````python
# Вариант решения с использованием таблицы сопряженности

import pandas as pd

data = {
    'Name': ['Анна', 'Антон', 'Мария', 'Макс', 'Макс', 'Алекс', 'Антон', 'Мария', 'Макс', 'Алекс', 'Антон', 'Анна'],
    'Game': ['тетрис', 'змейка', 'змейка', 'тетрис', 'тетрис', 'змейка', 'тетрис', 'тетрис', 'тетрис', 'змейка', 'змейка', 'тетрис'],
    'Points': [145, 2524, 343, 4212, 5212, 6421, 745, 842, 524, 3224, 212, 1214]
}

df = pd.DataFrame(data)

# Таблица сопряженности — среднее количество очков по игроку и игре
table = pd.crosstab(index=df["Name"],
                    columns=df["Game"],
                    values=df["Points"],
                    aggfunc="mean")

print("Среднее количество очков каждого игрока по каждой игре (через таблицу сопряженности):")
print(table)
````

**2. Построение диаграмм**

Напишите программу для анализа данных по потреблению электроэнергии в 25-квартирном доме в текущем месяце. Нужно:
1. Сгенерировать случайные данные о потреблении каждой квартиры – от 100 до 1000 кВт.
1. Визуализировать на круговой диаграмме энергопотребление каждой квартиры.
1. Представить ту же информацию в виде столбчатой диаграммы.



````python
import random
import matplotlib.pyplot as plt

# 1. Генерируем данные: 25 квартир, потребление от 100 до 1000 кВт
num_flats = 25

flats = []
consumption = []

for i in range(num_flats):
    flats.append("Кв " + str(i + 1))
    consumption.append(random.randint(100, 1000))

# 2. Круговая диаграмма
plt.figure(figsize=(8, 8))
plt.pie(consumption, labels=flats, autopct='%1.1f%%', startangle=90)
plt.title("Потребление электроэнергии по квартирам (круговая диаграмма)")
plt.axis('equal')  # чтобы круг не был овалом
plt.show()
````

````python
# 3. Столбчатая диаграмма
plt.figure(figsize=(12, 6))
plt.bar(flats, consumption)
plt.title("Потребление электроэнергии по квартирам (столбчатая диаграмма)")
plt.xlabel("Квартиры")
plt.ylabel("Потребление, кВт")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
````
